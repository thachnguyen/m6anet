```python
import pandas as pd
```

```python
nanom6a_results = "/mnt/dataHDD/chris/GohGIS_Hek293T_directRNA_Rep2/result_final/ratio.0.5.tsv"
df = []
with open(nanom6a_results) as f:
    for line in f:
        strip_line = line.strip("\n")
        strip_line = strip_line.split("\t")
        tx = strip_line[0].split("|")[0]
        for sub_line in strip_line[1:]:
            tx_pos, mod_number, total_number, ratio = sub_line.split("|")
            df += [(tx, tx_pos, mod_number, total_number, ratio)]
```

```python
df = pd.DataFrame(df, columns=["transcript_id", "coordinate", "n_mod", "total", "fraction_modified"])
```

```python
df["transcript_position"] = df["coordinate"].astype('int') - 1
```

```python
df.to_csv("~/nanom6a.csv.gz", index=False)
```
